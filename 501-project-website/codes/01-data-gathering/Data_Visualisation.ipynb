```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('classic')
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import seaborn as sns
import pandas as pd
```

```python
api_data_2021 = pd.read_csv('DC_2020.csv')
api_data_2020 = pd.read_csv('DC_2021.csv')
air_data = pd.read_csv('air_data_DC.csv')
forecast= pd.read_csv('Forecast_new.csv')

```

```python

from re import X
sns.boxplot(x="Year", y = "AvgMaxtemp" , data = forecast, palette="Paired")
sns.set_style("whitegrid")
#label the graph
plt.title("Average Max Temperature in Washington DC over the years")
plt.xlabel("Year")
plt.ylabel("Average Max Temperature")
plt.show()


```

```python
sns.lineplot(x="Year", y = "Ozone" , data = forecast, color='crimson', linewidth=1.5)
sns.set_style("whitegrid")
#label the graph
plt.title("Average Ozone pollutant in Air Washington DC over the years")
plt.xlabel("Year")
plt.ylabel("Average Ozone pollutant")
```

```python
f, ax = plt.subplots(1,2, figsize=(15, 6), dpi = 100)
sns.histplot(api_data_2020['tempmax'], kde=True, bins=20, color='purple',linewidth = 1, edgecolor='crimson', ax=ax[0])
ax[0].set_title('Temperature Variation in 2020', fontsize=15)
ax[0].set_xlabel('Maximum Temperature', fontsize=10)
ax[0].set_ylabel('Variation across the year', fontsize=10)


sns.histplot(api_data_2021['tempmax'], kde=True, bins=20, color='purple',linewidth = 1, edgecolor='crimson', ax=ax[1])
ax[1].set_title('Temperature Variation in 2020')
ax[1].set_xlabel('Maximum Temperature')
ax[1].set_ylabel('Variation across the year')

plt.show()

sns.set_style("whitegrid")
```

```python
sns.scatterplot(x="Date", y = "DAILY_AQI_VALUE" , data = air_data, color='crimson', linewidth=1.5)
sns.set_style("whitegrid")
#label the graph
plt.title("Air Quality Index in Washington DC in past two years")
plt.xlabel("Date from 2020 to 2022")
plt.ylabel("Air Quality Index")
```

```python
fx, ax = plt.subplots(1,2, figsize=(15, 6), dpi = 100)
sns.set_style("whitegrid")
sns.lineplot(x="precipprob", y = "precip" , data = api_data_2020, color='crimson', linewidth=1.5, ax=ax[0])
ax[0].set_title('Precipitation Probability in 2020', fontsize=15)
ax[0].set_xlabel('Precipitation Probability', fontsize=10)
ax[0].set_ylabel('Humidity', fontsize=10)

sns.set_style("whitegrid")
sns.lineplot(x="precipprob", y = "precip" , data = api_data_2021, color='crimson', linewidth=1.5, ax=ax[1])
ax[1].set_title('Precipitation Probability in 2021', fontsize=15)
ax[1].set_xlabel('Precipitation Probability', fontsize=10)
ax[1].set_ylabel('Humidity', fontsize=10)
```
